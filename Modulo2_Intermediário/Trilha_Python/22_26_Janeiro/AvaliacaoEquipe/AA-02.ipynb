# Exercício 1

Com base nas dimensões de cada diamante se pode calcular o volume aproximado, considerando que eles como se fossem caixas. Adicione então uma nova coluna que mostre o preço por milímetro cubico de cada diamante. 


```python
#Fazer aqui o exercício 1

# Calcular o volume para cada diamante
diamonds['Volume (mm^3)'] = diamonds['x'] * diamonds['y'] * diamonds['z']

# Calcular o preço por milímetro cúbico
diamonds['Preço por mm^3'] = (diamonds['price'] / diamonds['Volume (mm^3)']) * 1000

# Mostrar o DataFrame resultante
print(diamonds.head)
```

Utilizando os recursos do __Pandas__ mostre em um __Series__ o preço médio dos diamantes para cada tipo de corte corte. 

```python
# Fazer aqui o exercício 2

diamonds.groupby('cut')
precoMedio = diamonds.groupby('cut')['price'].median()
print("\nPreço Médio:\n")
print(precoMedio)
```

# Exercício 3

Utilizando os datasets anteriores identifique:
1. Quantos alunos de sexo masculino e feminino tem quem cada dataset. 
2. Qual a média final dos alunos em cada disciplina.
3. Qual a media final dos alunos cujos parentes (ao menos um) tem nível superior comparada à dos alunos em que nenhum dos pais tem esse tipo de formação.  

```python
# 1. Quantos alunos de sexo masculino e feminino tem quem cada dataset. 
count_math = math["sex"].value_counts()
count_port = port["sex"].value_counts()

print("Valores turma matematica:\n",count_math)
print("\nValore turma português\n",count_port)
```

```python
# 2. Qual a média final dos alunos em cada disciplina.

media_math = math["final_grade"].mean()
media_port = port["final_grade"].mean()

print("Resultado da média em matematica:")
print(media_math)
print("\nResultado da média em português:")
print(media_port)
```

```python
#3- Qual a media final dos alunos cujos parentes (ao menos um) tem nível superior comparada à dos alunos em que nenhum dos pais tem esse tipo de formação.


math['parents_higher_education'] = (math['mother_education'] == 'higher education') | (math['father_education'] == 'higher education')

# Separando os alunos com pelo menos um dos pais com nível superior e sem nenhum dos pais com nível superior
higher_edu_students = math[math['parents_higher_education']]
no_higher_edu_students = math[~math['parents_higher_education']]

# Calculando a média final para ambos os grupos
mean_final_grade_higher_edu = higher_edu_students['final_grade'].mean()
mean_final_grade_no_higher_edu = no_higher_edu_students['final_grade'].mean()

# Exibindo os resultados
print("Média final dos alunos de matemática cujos pais têm nível superior:", mean_final_grade_higher_edu)
print("Média final dos alunos de matemática cujos pais não têm nível superior:", mean_final_grade_no_higher_edu)

# Turma de português

port['parents_higher_education'] = (port['mother_education'] == 'higher education') | (port['father_education'] == 'higher education')

# Separando os alunos com pelo menos um dos pais com nível superior e sem nenhum dos pais com nível superior
higher_edu_students = port[port['parents_higher_education']]
no_higher_edu_students = port[~port['parents_higher_education']]

# Calculando a média final para ambos os grupos
mean_final_grade_higher_edu = higher_edu_students['final_grade'].mean()
mean_final_grade_no_higher_edu = no_higher_edu_students['final_grade'].mean()

# Exibindo os resultados
print("\nMédia final dos alunos de português cujos pais têm nível superior:", mean_final_grade_higher_edu)
print("Média final dos alunos de português cujos pais não têm nível superior:", mean_final_grade_no_higher_edu)
```
